```python
### READING DATA FROM CSV FILE
import pandas as pd
import numpy as np
df=pd.read_csv("orders.csv")
df.head(20)
```

```python
df["Ship Mode"].unique()
```

```python
df["Ship Mode"]=df["Ship Mode"].replace(["Not Available","unknown"],np.nan)
```

```python
df["Ship Mode"].unique()
```

```python
### RENAMING COLUMN NAMES TO LOWER CASE AND REPLACING SPACE TO UNDERSCORE
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
```

```python
df.columns  ### COLUMNS HAVE BEEN CHANGED
```

```python
df.head(5)
```

```python
### DERIVING NEW COLUMNS:- DISCOUNT, SELLING_PRICE, PROFUT
df["discount"]=df["list_price"]*df["discount_percent"]*.01
df["selling_price"]=df["list_price"]-df["discount"]
df["profit"]=df["selling_price"]-df["cost_price"]
```

```python
df
```

```python
df["order_date"]=pd.to_datetime(df["order_date"],format="%Y-%m-%d")
```

```python
df.head()
```

```python
### CHECKING DTYPES
df.dtypes
```

```python
### DROPPING COST_PRICE, LIST_PRICE,, DISCOUNT_PERCENT COLUMNS
df.drop(columns=["cost_price","list_price","discount_percent"],inplace=True)
```

```python
### LOADING DATA INTO MYSQL 
import sqlalchemy as sal
import mysql.connector
from sqlalchemy import create_engine,text
server_url="mysql+mysqlconnector://Parav_project:tony_stark@localhost:3306"
engine = create_engine(server_url)  ### CREATE ENGINE initializes a new SQLAlchemy engine instance that will manage connections to server in this case
with engine.connect() as connection:
    connection.execute(text("CREATE DATABASE IF NOT EXISTS ordersDB"))
    print("Database orders created")

db_engine = create_engine(server_url + "/ordersDB")    ### in this db_engine will manage connection to database
df.to_sql("orders", con=db_engine, if_exists="replace", index=False)
print("DataFrame successfully added to 'ordersDB' as the 'orders' table.")

```
